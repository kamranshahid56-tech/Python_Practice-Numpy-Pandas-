```python
circle = np.zeros((100,100))
```

```python
# Generate coordinate grids for rows (i) and columns (j)
# np.ogrid[:100, :100] creates two arrays:
# i is a column vector (100,1) and j is a row vector (1,100)
i, j = np.ogrid[:100,:100]
```

```python
center_i, center_j = 50,50
radius = 20
```

```python
# The Logic: Pythagorean theorem (i-50)^2 + (j-50)^2 < 20^2
# This creates a Boolean Mask (True inside the circle, False outside)
mask = (i-center_i)**2 + (j-center_j)**2 < radius**2
```

```python
circle[mask] = 255
```

```python
plt.imshow(circle, cmap = 'gray')
plt.title('Circle Generated via NumPy Logic')
plt.show()
```

```python
#Simple and High complexity
prime_numbers = []
for i in range(len(arr)):
  x = arr[arr[i] % arr == 0]
  if len(x) == 1:
    prime_numbers.append(int(arr[i]))
  else:
    print(f"{arr[i]} is not a prime number")
print(prime_numbers)
```

```python
#Efficient and Low Complexity (Sieve Logic)
# 1. Create a boolean array: Assume everything is prime at first
n = 50
is_prime = np.ones(n + 1, dtype=bool)

# 0 and 1 are not prime
is_prime[0] = is_prime[1] = False

# 2. The Sieve Logic
for p in range(2, int(np.sqrt(n)) + 1):
    if is_prime[p]:
        # Vectorized Masking:
        # Start at p*p, go to the end, step by p
        # Set all these multiples to False
        is_prime[p*p : n+1 : p] = False

# 3. Use the mask to get the numbers
primes = np.nonzero(is_prime)[0]

print("Prime numbers up to 50:")
print(primes)
```

```python
#Oil & Gas
import numpy as np
def summary(lists):
  total = np.sum(lists)
  avg = np.mean(lists)
  print(f'Total : {total:,.2f}')
  print(f'Average : {avg:,.2f}')
  #by using return, I can use these variables outside function as well
  return total, avg


data = [2.4,2.45,2.7,2.45,6,12,63,22,10020040]
total_val, avg_val = summary(data)

mid = total_val/2
print(mid)
```

```python
class summary:
  def __init__(self,lists):
    self.lists = lists
  def summary(self):
    total = np.sum(self.lists)
    avg = np.mean(self.lists)
    print(f'Total : {total:,.2f}')
    print(f'Average : {avg:,.2f}')
    return total, avg

summary1 = summary(data)
total_value,avg_value = summary1.summary()

print(total_value)
```

```python
#AlexNet Shape Validator
def shape_validator(y_labels):
  if y_labels.ndim == 2:
    return y_labels
  else:
    return y_labels.reshape(-1,1)

y_labels = np.array([1,2,3,4,5])
print(y_labels)
print(f'Original Shape is {y_labels.shape}')
y_labels = shape_validator(y_labels)
print(y_labels)
print(f'Modified Shape is {y_labels.shape}')
```

```python
from numpy._core.fromnumeric import reshape
class ShapeValidatorClass:
  def __init__(self,y_labels):
    self.y_labels = y_labels
  def shape_validator(self):
    if self.y_labels.ndim == 2:
      return self.y_labels
    else:
      return self.y_labels.reshape(-1,1)

y_labels = np.array([1,2,3,4,5])
print(y_labels)
print(f'Original Shape is {y_labels.shape}')
validator = ShapeValidatorClass(y_labels)
y_labels = validator.shape_validator()
print(y_labels)
print(f'Modified Shape is {y_labels.shape}')
```

```python
def clean_phone(phone):
  digits = re.sub(r'\D','',str(phone))
  if len(digits) == 10:
    return f"{digits[:3]}-{digits[3:6]}-{digits[6:]}"
  return None

def process_leads(input_file, output_file):
  df = pd.read_csv(input_file)
  initial_count = len(df)

# 2. Basic Validation: Email must contain '@' and '.'
# This also handles the 'not-an-email' cases
  df = df[df['email'].str.contains(r'[^@]+@[^@]+\.[^@]+', na = False)].copy()

  df['email'] = df['email'].str.lower()
  df['phone'] = df['phone'].apply(clean_phone)

  #Duplication
  df['signup_date'] = pd.to_datetime(df['signup_date'])
  df = df.sort_values(by = 'signup_date')
  df = df.drop_duplicates(subset = 'email', keep = 'last')

  df = df.dropna(subset=['phone'])
  df.to_csv(output_file, index=False)

  print("--Processing Complete")
  print(f"Total leads Count: {initial_count}")
  print(f"Final leads Count: {len(df)}")
  print(f"Removed/Invalid:     {initial_count - len(df)}")
  print(f"Saved to: {output_file}")

process_leads('raw_leads.csv', 'master_leads.csv')

```
