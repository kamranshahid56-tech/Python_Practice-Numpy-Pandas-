```python
circle = np.zeros((100,100))
```

```python
# Generate coordinate grids for rows (i) and columns (j)
# np.ogrid[:100, :100] creates two arrays:
# i is a column vector (100,1) and j is a row vector (1,100)
i, j = np.ogrid[:100,:100]
```

```python
center_i, center_j = 50,50
radius = 20
```

```python
# The Logic: Pythagorean theorem (i-50)^2 + (j-50)^2 < 20^2
# This creates a Boolean Mask (True inside the circle, False outside)
mask = (i-center_i)**2 + (j-center_j)**2 < radius**2
```

```python
circle[mask] = 255
```

```python
plt.imshow(circle, cmap = 'gray')
plt.title('Circle Generated via NumPy Logic')
plt.show()
```
