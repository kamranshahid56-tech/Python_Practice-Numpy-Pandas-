```python
circle = np.zeros((100,100))
```

```python
# Generate coordinate grids for rows (i) and columns (j)
# np.ogrid[:100, :100] creates two arrays:
# i is a column vector (100,1) and j is a row vector (1,100)
i, j = np.ogrid[:100,:100]
```

```python
center_i, center_j = 50,50
radius = 20
```

```python
# The Logic: Pythagorean theorem (i-50)^2 + (j-50)^2 < 20^2
# This creates a Boolean Mask (True inside the circle, False outside)
mask = (i-center_i)**2 + (j-center_j)**2 < radius**2
```

```python
circle[mask] = 255
```

```python
plt.imshow(circle, cmap = 'gray')
plt.title('Circle Generated via NumPy Logic')
plt.show()
```

```python
arr = np.arange(2,50)
print(arr)
```

```python
#Simple and High complexity
prime_numbers = []
for i in range(len(arr)):
  x = arr[arr[i] % arr == 0]
  if len(x) == 1:
    prime_numbers.append(int(arr[i]))
  else:
    print(f"{arr[i]} is not a prime number")
print(prime_numbers)
```

```python
#Efficient and Low Complexity (Sieve Logic)
# 1. Create a boolean array: Assume everything is prime at first
n = 50
is_prime = np.ones(n + 1, dtype=bool)

# 0 and 1 are not prime
is_prime[0] = is_prime[1] = False

# 2. The Sieve Logic
for p in range(2, int(np.sqrt(n)) + 1):
    if is_prime[p]:
        # Vectorized Masking:
        # Start at p*p, go to the end, step by p
        # Set all these multiples to False
        is_prime[p*p : n+1 : p] = False

# 3. Use the mask to get the numbers
primes = np.nonzero(is_prime)[0]

print("Prime numbers up to 50:")
print(primes)
```

```python
n = 25
is_prime = np.ones(n+1, dtype = bool)
is_prime[0] = is_prime[1] = False

for p in range(2, int(np.sqrt(n))+1):
  if is_prime[p]:
    is_prime[p*p : n+1: p] = False
primes = np.nonzero(is_prime)[0]
print("Prime numbers up to 50:")
print(primes)

```

```python
import pandas as pd

data = {
    'Month': ['Jan', 'Jan', 'Feb', 'Feb'],
    'City': ['NY', 'LA', 'NY', 'LA'],
    'Revenue': [100, 150, 200, 80],
    'Year': [2019, 2020, 2019, 2020]
}
df = pd.DataFrame(data)

# Logic: Since NY has two entries for 2019 (Jan and Feb),
# we tell Pandas to SUM them together.
pivot_df = df.pivot_table(index='City', columns='Year', values='Revenue', aggfunc='sum')

print('Pivoted DataFrame (Summed Revenue per Year):')
print(pivot_df)
```

```python
# We need to reset the index first because 'City' is currently the row label
long_df = pivot_df.reset_index().melt(id_vars= 'City', var_name='Year', value_name='Revenue')
```

```python
import pandas as pd

# Table 1: Employees (We have two Alices!)
employees = pd.DataFrame({
    'Name': ['Alice', 'Alice', 'Bob'],
    'Dept': ['IT', 'HR', 'IT'],
    'ID': [101, 102, 103]
})

# Table 2: Salaries (Salaries vary by both Name and Dept)
salaries = pd.DataFrame({
    'Name': ['Alice', 'Alice', 'Bob'],
    'Dept': ['IT', 'HR', 'IT'],
    'Salary': [70000, 50000, 80000]
})

# THE LOGIC: Merge using a LIST of columns
# This ensures Alice in IT gets 70k and Alice in HR gets 50k
merged_df = pd.merge(employees, salaries, on=['Name', 'Dept'])
merged_df1 = pd.merge(employees, salaries, on='Name')

print(merged_df1)
print(merged_df)
```
